## Write function to reverse the string in python

```python
from typing import List 


def reverseString(string: str)->str:
    return string[::-1]

def reverseString2(string: str)->str:
    l=0
    r=len(string)-1
    while(l<r):
        string[l],string[r]=string[r],string[l]
        l+=1
        r+=1
    
    return string

input="ABCD"
print(input)
print(reverseString2(input))
```
