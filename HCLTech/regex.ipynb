```python
import re

line = '2025-12-29 19:54:10.123 INFO auth user=chinmaey ip=10.2.3.4 latency_ms=37 msg="login ok"'

pattern = re.compile(
    r'^(?P<ts>\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}\.\d{3})\s+'
    r'(?P<level>INFO|WARN|ERROR|DEBUG)\s+'
    r'(?P<module>\w+)\s+'
    r'user=(?P<user>\S+)\s+'
    r'ip=(?P<ip>\S+)\s+'
    r'latency_ms=(?P<latency_ms>\d+)\s+'
    r'msg="(?P<msg>.*)"$'
)

match = pattern.match(line)
if match:
    print(match.groupdict())
```
